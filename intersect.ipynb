```python
from datetime import datetime
import geopandas as gpd
```

```python
# Load the two shapefiles
forest = gpd.read_file(r"C:\Users\taohuang\Downloads\BdyAdm_LSRS_AdministrativeForest\S_USA.BdyAdm_LSRS_AdministrativeForest.shp")   # polygon(s) you want to subset
states = gpd.read_file(r"C:\Users\taohuang\Downloads\WA_OR_CA.gpkg")   # reference polygon(s)

# print( [forest.crs, states.crs])
if forest.crs != states.crs:
    states = states.to_crs(forest.crs)
```

```python
# Polygons from gdf1 that touch or overlap any polygon in gdf2
selected = forest[forest.intersects(states.union_all())].copy()

selected.to_file("westcoast_forests.shp")
```
