```python
import geopandas as gpd
import requests
from zipfile import ZipFile
from io import BytesIO
```

```python

# Read TIGER state boundaries directly from Census HTTPS (2023 or newer)
#url = "https://www2.census.gov/geo/tiger/TIGER2024/STATE/tl_2024_us_state.zip"
url = "https://www2.census.gov/geo/tiger/TIGER2025/STATE/tl_2025_us_state.zip"
states = gpd.read_file(url)

# Filter WA, OR, CA
#west_coast = states[states.STUSPS.isin(['WA', 'OR', 'CA'])].dissolve().geometry.iloc[0]
WA_OR_CA = states[states.STUSPS.isin(['WA', 'OR', 'CA'])]

WA_OR_CA.to_file("WA_OR_CA.gpkg", driver="GPKG")
```
