```python
import pandas as pd
import os
# List all files
os.listdir()
```

```python
# path to the Excel file
xlsx_path = "S.xlsx"

# read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(xlsx_path, sheet_name=None)

dfs
```

```python
s_1 = "wildfires"
s_1_df = dfs[ s_1 ]
indicator_counts_df_1 = s_1_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_1.columns = ['Indicator', 'Count']
indicator_counts_df_1['Susceptibility'] =  s_1
indicator_counts_df_1= indicator_counts_df_1[[      "Susceptibility" , "Indicator",  "Count" ]]
print(indicator_counts_df_1)
indicator_counts_df_1.to_csv("S_indicator_counts_wildfires.csv")
```

```python
s_2 = "droughts"
s_2_df = dfs[ s_2 ]
indicator_counts_df_2 = s_2_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_2.columns = ['Indicator', 'Count']
indicator_counts_df_2['Susceptibility'] =  s_2
indicator_counts_df_2= indicator_counts_df_2[[      "Susceptibility" , "Indicator",  "Count" ]]
print(indicator_counts_df_2)
indicator_counts_df_2.to_csv("S_indicator_counts_drought.csv")
```

```python
s_3 = "insect outbreaks"
s_3_df = dfs[ s_3 ]
indicator_counts_df_3 = s_3_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_3.columns = ['Indicator', 'Count']
indicator_counts_df_3['Susceptibility'] =  s_3
indicator_counts_df_3= indicator_counts_df_3[[      "Susceptibility" , "Indicator",  "Count" ]]
print(indicator_counts_df_3)
indicator_counts_df_3.to_csv("S_indicator_counts_insect.csv")
```

```python

S_indicator_counts = pd.concat([ indicator_counts_df_1,  indicator_counts_df_2, indicator_counts_df_3], ignore_index=True)

print(S_indicator_counts)

S_indicator_counts.to_csv("S_indicator_counts.csv")
```

```python
#s_1 = "wildfires"
#s_1_df = dfs[ s_1 ]

data_counts_df_1 = s_1_df['Dataset (consistent name)'].value_counts().reset_index()
data_counts_df_1.columns = ['Dataset', 'Count']
data_counts_df_1['Susceptibility'] =  s_1
data_counts_df_1=  data_counts_df_1[[      "Susceptibility" , "Dataset",  "Count" ]]
print( data_counts_df_1 )
data_counts_df_1.to_csv("S_data_counts_wildfire.csv")
```

```python
data_counts_df_2 = s_2_df['Dataset (consistent name)'].value_counts().reset_index()
data_counts_df_2.columns = ['Dataset', 'Count']
data_counts_df_2['Susceptibility'] =  s_2
data_counts_df_2=  data_counts_df_2[[      "Susceptibility" , "Dataset",  "Count" ]]
print( data_counts_df_2 )
data_counts_df_2.to_csv("S_data_counts_drought.csv")
```

```python
data_counts_df_3 = s_3_df['Dataset (consistent name)'].value_counts().reset_index()
data_counts_df_3.columns = ['Dataset', 'Count']
data_counts_df_3['Susceptibility'] =  s_3
data_counts_df_3=  data_counts_df_3[[      "Susceptibility" , "Dataset",  "Count" ]]
print( data_counts_df_3 )
data_counts_df_3.to_csv("S_data_counts_insect.csv")
```

```python
S_data_counts = pd.concat([ data_counts_df_1,  data_counts_df_2, data_counts_df_3], ignore_index=True)

print(S_data_counts)

S_data_counts.to_csv("S_data_counts.csv")
```

```python
s_1_df_comb = s_1_df[['Indicator (consistent name)', 'Dataset (consistent name)']].drop_duplicates()
s_1_df_comb.columns = ['Indicator', 'Dataset']

print(s_1_df_comb)

# Merge with the count data (assuming data_counts_df_1 exists)
s_1_df_ind_data_ct = pd.merge(
    s_1_df_comb, 
    data_counts_df_1, 
    on='Dataset', 
    how='left'
)

print("s_1_df_ind_data_ct:", s_1_df_ind_data_ct)

# Sort by count descending
s_1_df_ind_data_ct_sort = s_1_df_ind_data_ct.sort_values(by='Count', ascending=False)

print("s_1_df_ind_data_ct_sort:", s_1_df_ind_data_ct_sort)

# Save to CSV with updated name
s_1_df_ind_data_ct_sort.to_csv("S_ind_data_counts_fire.csv", index=False)
```

```python
# Create the unique Indicator-Dataset combinations
s_2_df_comb = s_2_df[['Indicator (consistent name)', 'Dataset (consistent name)']].drop_duplicates()
s_2_df_comb.columns = ['Indicator', 'Dataset']

print(s_2_df_comb)

# Merge with the count data (assuming data_counts_df_2 exists)
s_2_df_ind_data_ct = pd.merge(
    s_2_df_comb, 
    data_counts_df_2, 
    on='Dataset', 
    how='left'
)

print("s_2_df_ind_data_ct:", s_2_df_ind_data_ct)

# Sort by count descending
s_2_df_ind_data_ct_sort = s_2_df_ind_data_ct.sort_values(by='Count', ascending=False)

print("s_2_df_ind_data_ct_sort:", s_2_df_ind_data_ct_sort)

# Save to CSV with updated name
s_2_df_ind_data_ct_sort.to_csv("S_ind_data_counts_drought.csv")
```
