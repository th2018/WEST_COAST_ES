```python
import pandas as pd
import os
os.listdir()
```

```python
# path to your Excel file
xlsx_path = "LAC.xlsx"

# read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(xlsx_path, sheet_name=None)
```

```python
hazard_1 = "wildfires"
hazard_1_df = dfs[ hazard_1 ]
indicator_counts_df_1 = hazard_1_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_1.columns = ['Indicator', 'Count']
indicator_counts_df_1['Hazard'] =  hazard_1
indicator_counts_df_1= indicator_counts_df_1[[      "Hazard" , "Indicator",  "Count" ]]
print(indicator_counts_df_1)
indicator_counts_df_1.to_csv("lac_ind_count_fire.csv")
```

```python
hazard_2 = "droughts"
hazard2_df = dfs[ hazard_2 ]
indicator_counts_df_2 = hazard2_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_2.columns = ['Indicator', 'Count']
indicator_counts_df_2['Hazard'] =  hazard_2
indicator_counts_df_2= indicator_counts_df_2[[      "Hazard" , "Indicator",  "Count" ]]
print(indicator_counts_df_2)
indicator_counts_df_2.to_csv("lac_ind_count_drought.csv")
```

```python
hazard_3 = "insect outbreaks"
hazard3_df = dfs[ hazard_3 ]
indicator_counts_df_3 =  hazard3_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_3.columns = ['Indicator', 'Count']
indicator_counts_df_3['Hazard'] =  hazard_3
indicator_counts_df_3= indicator_counts_df_3[[      "Hazard" , "Indicator",  "Count" ]]
print(indicator_counts_df_3)
indicator_counts_df_3.to_csv("lac_ind_count_insect.csv")
```

```python
LAC_indicator_counts = pd.concat([ indicator_counts_df_1,  indicator_counts_df_2, indicator_counts_df_3], ignore_index=True)
print(LAC_indicator_counts)
LAC_indicator_counts.to_csv("LAC_indicator_counts.csv")
```

```python
hazard_1 = "wildfires"
hazard_1_df = dfs[ hazard_1 ]
dataset_counts_df_1 = hazard_1_df['Dataset (consistent name)'].value_counts().reset_index()
dataset_counts_df_1.columns = ['Dataset', 'Count']
dataset_counts_df_1['Hazard'] =  hazard_1
dataset_counts_df_1= dataset_counts_df_1[[      "Hazard" , "Dataset",  "Count" ]]
print(dataset_counts_df_1)
dataset_counts_df_1.to_csv("lac_data_count_wildfires.csv")
```

```python
hazard_2 = "droughts"
hazard2_df = dfs[ hazard_2 ]
dataset_counts_df_2 = hazard2_df['Dataset (consistent name)'].value_counts().reset_index()
dataset_counts_df_2.columns = ['Dataset', 'Count']
dataset_counts_df_2['Hazard'] =  hazard_2
dataset_counts_df_2= dataset_counts_df_2[[      "Hazard" , "Dataset",  "Count" ]]
print(dataset_counts_df_2)
dataset_counts_df_2.to_csv("lac_data_count_drought.csv")
```

```python
hazard_3 = "insect outbreaks"
hazard3_df = dfs[ hazard_3 ]
dataset_counts_df_3 =  hazard3_df['Dataset (consistent name)'].value_counts().reset_index()
dataset_counts_df_3.columns = ['Dataset', 'Count']
dataset_counts_df_3['Hazard'] =  hazard_3
dataset_counts_df_3= dataset_counts_df_3[[      "Hazard" , "Dataset",  "Count" ]]
print(dataset_counts_df_3)
dataset_counts_df_3.to_csv("lac_data_count_insect.csv")
```

```python
LAC_dataset_counts = pd.concat([ dataset_counts_df_3,  dataset_counts_df_2, dataset_counts_df_1], ignore_index=True)
print(LAC_dataset_counts)
LAC_dataset_counts.to_csv("LAC_dataset_counts.csv")
```

```python

s_1_df_comb = hazard_1_df[['Indicator (consistent name)', 'Dataset (consistent name)']].drop_duplicates()
s_1_df_comb.columns = ['Indicator', 'Dataset']

s_1_df_ind_data_ct = pd.merge(
    s_1_df_comb, 
    dataset_counts_df_1, 
    on='Dataset', 
    how='left'
)

s_1_df_ind_data_ct_sort = s_1_df_ind_data_ct.sort_values(by='Count', ascending=False)

print("s_1_df_ind_data_ct_sort:",s_1_df_ind_data_ct_sort)

s_1_df_ind_data_ct_sort.to_csv("LAC_ind_data_counts_wildfire.csv", index=False)
```

```python
s_2_df_comb = hazard2_df[['Indicator (consistent name)', 'Dataset (consistent name)']].drop_duplicates()
s_2_df_comb.columns= [ 'Indicator', 'Dataset']


s_2_df_ind_data_ct = pd.merge( s_2_df_comb, dataset_counts_df_2, on='Dataset', how='left')



s_2_df_ind_data_ct_sort = s_2_df_ind_data_ct.sort_values(by='Count', ascending=False)
print("s_2_df_ind_data_ct_sort:",s_2_df_ind_data_ct_sort)

s_2_df_ind_data_ct_sort.to_csv("LAC_ind_data_counts_drought.csv")
```

```python
s_3_df_comb = hazard3_df[['Indicator (consistent name)', 'Dataset (consistent name)']].drop_duplicates()
s_3_df_comb.columns= [ 'Indicator', 'Dataset']
print(s_3_df_comb)
s_3_df_ind_data_ct = pd.merge( s_3_df_comb, dataset_counts_df_3, on='Dataset', how='left')
print("s_3_df_ind_data_ct:",s_3_df_ind_data_ct)

s_3_df_ind_data_ct_sort = s_3_df_ind_data_ct.sort_values(by='Count', ascending=False)
print("s_3_df_ind_data_ct_sort:",s_3_df_ind_data_ct_sort)

s_3_df_ind_data_ct_sort.to_csv("LAC_ind_data_counts_insect.csv")
```
