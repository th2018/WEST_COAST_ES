```python
import pandas as pd
import os
# List all files
os.listdir()
```

```python
# path to the Excel file
xlsx_path = "H.xlsx"

# read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(xlsx_path, sheet_name=None)

dfs
```

```python
s_1 = "wildfires"
s_1_df = dfs[ s_1 ]
indicator_counts_df_1 = s_1_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_1.columns = ['Indicator', 'Count']
indicator_counts_df_1['Hazard'] =  s_1
indicator_counts_df_1= indicator_counts_df_1[[      "Hazard" , "Indicator",  "Count" ]]
print(indicator_counts_df_1)
indicator_counts_df_1.to_csv("H_indicator_counts_wildfire.csv")
```

```python
s_2 = "droughts"
s_2_df = dfs[ s_2 ]
indicator_counts_df_2 = s_2_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_2.columns = ['Indicator', 'Count']
indicator_counts_df_2['Hazard'] =  s_2
indicator_counts_df_2= indicator_counts_df_2[[      "Hazard" , "Indicator",  "Count" ]]
print(indicator_counts_df_2)
indicator_counts_df_2.to_csv("H_indicator_counts_drought.csv")
```

```python
s_3 = "insect outbreaks"
s_3_df = dfs[ s_3 ]
indicator_counts_df_3 = s_3_df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_3.columns = ['Indicator', 'Count']
indicator_counts_df_3['Hazard'] =  s_3
indicator_counts_df_3= indicator_counts_df_3[[      "Hazard" , "Indicator",  "Count" ]]
print(indicator_counts_df_3)
indicator_counts_df_3.to_csv("H_indicator_counts_insect.csv")
```

```python
H_indicator_counts = pd.concat([ indicator_counts_df_1,  indicator_counts_df_2, indicator_counts_df_3], ignore_index=True)

print(H_indicator_counts)

H_indicator_counts.to_csv("H_indicator_counts.csv")
```

```python
data_counts_df_1 = s_1_df['Dataset (consistent name)'].value_counts().reset_index()
data_counts_df_1.columns = ['Dataset', 'Count']
data_counts_df_1['Hazard'] =  s_1
data_counts_df_1=  data_counts_df_1[[      "Hazard" , "Dataset",  "Count" ]]
print( data_counts_df_1 )
data_counts_df_1.to_csv("H_data_counts_wildfires.csv")
```

```python
data_counts_df_2 = s_2_df['Dataset (consistent name)'].value_counts().reset_index()
data_counts_df_2.columns = ['Dataset', 'Count']
data_counts_df_2['Hazard'] =  s_2
data_counts_df_2=  data_counts_df_2[[      "Hazard" , "Dataset",  "Count" ]]
print( data_counts_df_2 )
data_counts_df_2.to_csv("H_data_counts_droughts.csv")
```

```python
data_counts_df_3 = s_3_df['Dataset (consistent name)'].value_counts().reset_index()
data_counts_df_3.columns = ['Dataset', 'Count']
data_counts_df_3['Hazard'] =  s_3
data_counts_df_3=  data_counts_df_3[[      "Hazard" , "Dataset",  "Count" ]]
print( data_counts_df_3 )


data_counts_df_3.to_csv("H_data_counts_insect.csv")
```

```python

H_data_counts = pd.concat([ data_counts_df_1,  data_counts_df_2, data_counts_df_3], ignore_index=True)

print(H_data_counts)

H_data_counts.to_csv("H_data_counts.csv")
```
