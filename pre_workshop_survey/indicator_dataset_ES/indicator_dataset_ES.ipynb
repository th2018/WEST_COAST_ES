```python
import pandas as pd
import os
# List all files
os.listdir()
```

```python
# path to the Excel file
es_xlsx_path = "indicator_dataset_ES.xlsx"

# read all sheets into a dictionary of DataFrames

dfs= pd.read_excel(es_xlsx_path, sheet_name=None)

dfs

#print(dfs['Sheet1'])
```

```python
s_1 = "carbon"
df=dfs[ s_1]
indicator_counts_df_1 = df['Indicator (consistent name)'].value_counts().reset_index()
indicator_counts_df_1.columns = ['Indicator', 'Count']
indicator_counts_df_1['Ecosystem service'] =  s_1
indicator_counts_df_1= indicator_counts_df_1[[      "Ecosystem service" , "Indicator",  "Count" ]]
print(indicator_counts_df_1)
```
