```python
import geopandas as gpd
```

```python
# Replace with the path to your input shapefile
input_shp = r'C:\Users\taohuang\Downloads\westcoast_forests.shp'

# Replace with the desired output path for the new shapefile
output_shp = r'C:\Users\taohuang\Downloads\westcoast_forests_explode.shp'

# Read the input shapefile
gdf = gpd.read_file(input_shp)
print( type(gdf) )

# Explode multi-part polygons (MultiPolygon) into individual single-part polygons
# This splits "untouched" (likely meaning multi-part or connected) polygons into separate features
exploded_gdf = gdf.explode(index_parts=True)

# Optional: Reset the index for a clean output
exploded_gdf = exploded_gdf.reset_index(drop=True)

# Save the result to a new shapefile
exploded_gdf.to_file(output_shp)

print(f"Split complete. Output saved to {output_shp}")
```
